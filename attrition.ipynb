## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
# 'You can use any 10 columns you'd like' - Challenge 19 Instructions
X_list = ['Age', 'Education', 'JobInvolvement', 'JobSatisfaction', 'NumCompaniesWorked', 'PerformanceRating', 'TotalWorkingYears', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsWithCurrManager']
display(len(X_list))

# Create X_df using your selected columns
X_df = attrition_df[X_list]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

#There is no selected X data that needs converted into a numerical datatype. All chosen columns chosen were numerical
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
ohedept = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
ohedept.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_dept_encoded = ohedept.transform(y_train[['Department']])
y_test_dept_encoded = ohedept.transform(y_test[['Department']])


y_train_dept_encoded
```

```python
# Create a OneHotEncoder for the Attrition column
oheatt = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
oheatt.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_att_encoded = oheatt.transform(y_train[['Attrition']])
y_test_att_encoded = oheatt.transform(y_test[['Attrition']])

y_train_att_encoded
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_columns = len(X_train.columns)

# Create the input layer
input_layer = layers.Input(shape=(num_columns,))

# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(units=32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
dept_output_layer = layers.Dense(3, activation='softmax', name='department_output')(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
att_hidden_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
att_output_layer = layers.Dense(2, activation='sigmoid', name='attrition_output')(att_hidden_layer)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output_layer, att_output_layer])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(X_train_scaled,
                    {'department_output': y_train_dept_encoded,
                     'attrition_output': y_train_att_encoded},
                    epochs=100)
```

```python
# Evaluate the model with the testing data
test_accuracy = model.evaluate(X_test_scaled,
              {'department_output': y_test_dept_encoded,
              'attrition_output': y_test_att_encoded},
              verbose=2)
```

```python
# Print the accuracy for both department and attrition
print(f'Department Predictions Accuracy: {test_accuracy[2]}')
print(f'Attrition Predictions Accuracy: {test_accuracy[1]}')
```
